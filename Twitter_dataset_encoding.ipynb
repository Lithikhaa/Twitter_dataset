```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('imdb_reviews.csv', error_bad_lines=False, engine ='python')
```

```python
df.head(10)
```

```python
df['lnum']=df.sentiment.map({'pos':1,'neg':0})
df
```

```python
x = df['text']
y = df['lnum']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
xtrain_dtm = cv.fit_transform(xtrain)
xtest_dtm=cv.transform(xtest)
print(cv.get_feature_names_out())
df=pd.DataFrame(xtrain_dtm.toarray(),columns=cv.get_feature_names_out())
```

```python
df
```

```python
from sklearn import metrics
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(xtrain_dtm,ytrain)
predicted = clf.predict(xtest_dtm)
```

```python
metrics.confusion_matrix(ytest,predicted)
```

```python
metrics.accuracy_score(ytest,predicted)
```

```python
print("Precision :",metrics.precision_score(ytest, predicted,average='micro'))
print("Recall : ",metrics.recall_score(ytest, predicted,average='micro'))
```
